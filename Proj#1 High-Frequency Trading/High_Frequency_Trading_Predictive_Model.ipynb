```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from google.colab import drive
import pandas as pd
pd.set_option('display.max_columns', None)

drive.mount('/content/drive')
```

```python
folder_path = "/content/drive/MyDrive/Project/training"
mat_files = [file for file in os.listdir(folder_path) if file.endswith('.mat')]

training = pd.DataFrame()

for mat_file in mat_files:

    file_path = os.path.join(folder_path, mat_file)

    mat_data = loadmat(file_path)

    MidPrice = pd.DataFrame(mat_data['MidPrice'],columns=['MidPrice'])
    datam = pd.DataFrame(mat_data['datam'],columns=['bid1px', 'bid1sz','bid2px', 'bid2sz','bid3px', 'bid3sz','bid4px', 'bid4sz','bid5px', 'bid5sz',
                                            'ask1px','ask1sz','ask2px','ask2sz','ask3px','ask3sz','ask4px','ask4sz','ask5px','ask5sz',
                                          'last_px','volume','cumulative_volume','open interest','turnover','cumulative_turnover' ])
    tm_exch = pd.DataFrame(mat_data['tm_exch'],columns=['hh','mm','ss','xxx'])
    y = pd.DataFrame(mat_data['y'],columns=['y'])
    result = pd.concat([tm_exch, datam,MidPrice,y], axis=1)
    date_str = mat_file.split('_')[0]
    date_datetime = datetime.strptime(date_str, '%Y%m%d')
    print(mat_file,result.shape)
    result['date'] = date_datetime.strftime('%Y-%m-%d')

    training = pd.concat([training, result], ignore_index=True)

print('Data Loaded')
```

```python
training['morning'] = ((training['hh'] >= 9) & (training['hh'] <= 11)).astype(int)
training['afternoon'] = ((training['hh'] >= 13) & (training['hh'] <= 14)).astype(int)
training['date'] = pd.to_datetime(training['date'])
training['weekday'] = training['date'].dt.dayofweek
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', "Friday"]
for i, day in enumerate(weekdays):
    # Note: Monday=0, hence aligning with weekday values
    training[day] = (training['weekday'] == i).astype(int)

# Drop the temporary 'weekday' column as it's no longer needed
training.drop('weekday', axis=1, inplace=True)
training.drop('Friday', axis=1, inplace=True)
training['pricediff'] = training['MidPrice'].diff(periods=40)
training['price_change_1'] = training['last_px'].diff(1)
training['price_change_5'] = training['last_px'].diff(5)
training['price_change_10'] = training['last_px'].diff(10)

# 3: Moving averages
# Simple Moving Average (SMA) for the last_px over different periods
training['sma_5'] = training['last_px'].rolling(window=5).mean()
training['sma_10'] = training['last_px'].rolling(window=10).mean()

# 4: Volume and turnover features
# Moving averages for volume and turnover to capture trends
training['volume_sma_5'] = training['volume'].rolling(window=5).mean()
training['turnover_sma_5'] = training['turnover'].rolling(window=5).mean()

# 5: Bid-Ask spreads and imbalances
# Calculate the spread between the best bid and ask prices
training['spread'] = training['ask1px'] - training['bid1px']

# Calculate the order book imbalance
training['imbalance'] = (training['bid1sz'] - training['ask1sz']) / (training['bid1sz'] + training['ask1sz'])

# 6: Non-linear transformations
# Log transformation of volume and turnover for normalization
training['log_volume'] = np.log(training['volume'] + 1)  # Adding 1 to avoid log(0)
training['log_turnover'] = np.log(training['turnover'] + 1)


training.dropna(subset=['pricediff'], inplace=True)
for i in range(1,6):
  training['size_pressure'+str(i)] = training['bid'+str(i)+'sz']- training['ask'+str(i)+'sz']
```

```python
# 1 Time Segmentation
test['morning'] = ((test['hh'] >= 9) & (test['hh'] <= 11)).astype(int)
test['afternoon'] = ((test['hh'] >= 13) & (test['hh'] <= 14)).astype(int)
test['date'] = pd.to_datetime(test['date'])
test['weekday'] = test['date'].dt.dayofweek
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', "Friday"]
for i, day in enumerate(weekdays):
    # Note: Monday=0, hence aligning with weekday values
    test[day] = (test['weekday'] == i).astype(int)

# Drop the temporary 'weekday' column as it's no longer needed
test.drop('weekday', axis=1, inplace=True)
test.drop('Friday', axis=1, inplace=True)
test['pricediff'] = test['MidPrice'].diff(periods=40)

# 2: Price change
test['price_change_1'] = test['last_px'].diff(1)
test['price_change_5'] = test['last_px'].diff(5)
test['price_change_10'] = test['last_px'].diff(10)

# 3: Moving averages
# Simple Moving Average (SMA) for the last_px over different periods
test['sma_5'] = test['last_px'].rolling(window=5).mean()
test['sma_10'] = test['last_px'].rolling(window=10).mean()

# 4: Volume and turnover features
# Moving averages for volume and turnover to capture trends
test['volume_sma_5'] = test['volume'].rolling(window=5).mean()
test['turnover_sma_5'] = test['turnover'].rolling(window=5).mean()

# 5: Bid-Ask spreads and imbalances
# Calculate the spread between the best bid and ask prices
test['spread'] = test['ask1px'] - test['bid1px']

# Calculate the order book imbalance
test['imbalance'] = (test['bid1sz'] - test['ask1sz']) / (test['bid1sz'] + test['ask1sz'])

# 6: Non-linear transformations
# Log transformation of volume and turnover for normalization
test['log_volume'] = np.log(test['volume'] + 1)  # Adding 1 to avoid log(0)
test['log_turnover'] = np.log(test['turnover'] + 1)

test.dropna(subset=['pricediff'], inplace=True)
for i in range(1,6):
  test['size_pressure'+str(i)] = test['bid'+str(i)+'sz']- test['ask'+str(i)+'sz']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming 'training' and 'test' are your DataFrame names and already include the new features.

# Separate features and target variable

X_train = training.drop(['y','date'], axis=1)
X_test = test.drop(['y','date'], axis=1)
y_train = training['y']
y_test = test['y']


# Standardize the features (important for linear models)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the models
linear_model = LinearRegression()
ridge_model = Ridge(alpha= 0.1 )  # Alpha is a hyperparameter for Ridge; adjust as needed
lasso_model = Lasso(alpha= 0.1 )

# Train the models
linear_model.fit(X_train_scaled, y_train)
ridge_model.fit(X_train_scaled, y_train)
lasso_model.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_linear = linear_model.predict(X_test_scaled)
y_pred_ridge = ridge_model.predict(X_test_scaled)
y_pred_lasso = lasso_model.predict(X_test_scaled)

# Evaluate the models
print("Linear Regression Metrics:")
print("MSE:", mean_squared_error(y_test, y_pred_linear))
print("R^2:", r2_score(y_test, y_pred_linear))

print("\nRidge Regression Metrics:")
print("MSE:", mean_squared_error(y_test, y_pred_ridge))
print("R^2:", r2_score(y_test, y_pred_ridge))

print("\nLasso Regression Metrics:")
print("MSE:", mean_squared_error(y_test, y_pred_lasso))
print("R^2:", r2_score(y_test, y_pred_lasso))
```
